```python
# Import random package
import random

#Asking the user name
user_name= input(prompt="⚔ Insert your name ")
#Optional character skin
while True:
    user_skin= input(prompt=f"""♛ Choose an apparel
    ① (̶◉͛‿◉̶) 
    ② ✿◕ ‿ ◕✿ 
    ③ ಥ_ಥ 
    """)
    skins={"1":"(̶◉͛‿◉̶)","2":"✿◕ ‿ ◕✿","3":"ಥ_ಥ"}
    if user_skin in skins:
        apparel=skins[user_skin]
        break
    
    else:
        print("►I didn't get that.")
        continue
        

#Description of game theory situation by parts
print(f"""
► {user_name}, Welcome to 🅳🅸🆂🅱🅾🅰🆁🅳: a world where loyalty is as certain as fate.

After attempting to steal the 𝕾𝖔𝖚𝖑 𝖌𝖊𝖒 ❂ from its shrine,
you were caught by Azrael, the Holy Guardian of the Temple. 

However, you were not alone: Chlammy Zell  helped you despite
your disfuctional relationship. There are no other witnesses of your
crime.

To protect the Church's reputation in Disboard, Azrael
decided to judge you against each other.
""")
print(input(prompt="Press 🅴🅽🆃🅴🆁\n"))
print(f"""
►You were placed on the left wing of the Temple
meanwhile Chlammy is interrogated in the secret library.

The silence is broken by the slow steps of the Holy Guardian
getting closer to the room. 
""")
print(input(prompt="Press 🅴🅽🆃🅴🆁\n"))
print(f"""
►The door is opened and Azrael appears with a
mocking smile in his angelical face.
""")
print(input(prompt="Press 🅴🅽🆃🅴🆁\n"))
print(f"""
🗯️ 𝓐𝔃𝓻𝓪𝓮𝓵: The time has come, {user_name}.
Your sinner accomplice has already made her decision.
However, because our Lord's mercy is beyond measure,
you seem to have another opportunity.
""")
print(input(prompt="Press 🅴🅽🆃🅴🆁\n"))
print(f"""
►You have the chance of exchange roles with Fil Nilvalen,
a follower of Chlammy. She can take your place in the
judgment by claimming she used a mutation spell to
trick Azrael.
""")

#Defining the game mode by asking the user if wants to play or not.
#Using while True to avoid wrong answers/spells
while True:
        user_participation=input(prompt=f"""
        ►What will you do?
        
        {apparel+user_name}: I want to
        ① 𝑃𝑙𝑎𝑦
        ② 𝐿𝑒𝑡 𝐹𝑖𝑙 𝑡𝑎𝑘𝑒 𝑚𝑦 𝑝𝑙𝑎𝑐𝑒\n""")
        #Message is wants to play. Explaining possible outcomes.
        if user_participation=="1":
            #Explaining scenarios
            print(f"""
            🗯️ 𝓐𝔃𝓻𝓪𝓮𝓵: {user_name}. Before taking a decision, consider your options:
            
            Chlammy                                 You                
            {"_"*60}
                                         Unvoice                       Betray
                                         
                                 Both 2 year in jail              You're free
            Unvoice                                                     
                                                                               Chlammy's jailed 10 years
            {"_"*60}
                                 You're jailed 10 years       Both are jailed 7 years
            Betray   
                                 Chlammy is free
           
           🗯️ 𝓐𝔃𝓻𝓪𝓮𝓵: Now confess, shameless sinner.
            """)
                
            
            break
            
            #If does not want to play, scenarios where two Python players play. Explaning possible outcomes
        elif user_participation=="2":
            print(f"""
            🗯️ 𝓐𝔃𝓻𝓪𝓮𝓵: I was tricked! Bring here that shameless fool!
            
            ►Fil appears in the room, but Azrael refuses to totally free you.
            
            🗯️ 𝓐𝔃𝓻𝓪𝓮𝓵: I will withhold you until the final decision is made!
            Sinner Fil, consider your options carefully:
            
            Chlammy                                 Fil                
            {"_"*60}
                                         Unvoice                       Betray
              {"_"*60}                             
                                    Both 2 year in jail              Fil's free
            Unvoice                                                     
                                                                                  Chlammy's jailed 10 years
            {"_"*60}
                                    Fil's jailed 10 years            Both are jailed 7 years
            Betray   
                                    Chlammy is free
           
            """)
            print(input(prompt="Press 🅴🅽🆃🅴🆁\n"))
            
            
            break
            #When the incorrect format is typed.
        else:
            print(f"""
            🗯️ 𝓐𝔃𝓻𝓪𝓮𝓵: What is that supposed to mean?
           """)
            continue
            
#Starting the game with game mode defined previously
if user_participation=="1":
    #Using while True to avoid wrong answers/spells
    while True:
        user_decision=input(prompt=(f"""
        ►What will you do?
        
        {apparel+user_name}: I will
        ① 𝑅𝑒𝑚𝑎𝑖𝑛 𝑠𝑖𝑙𝑒𝑛𝑡
        ② 𝐵𝑒𝑡𝑟𝑎𝑦 𝐶ℎ𝑙𝑎𝑚𝑚𝑦
        """))
        #Pc player making a 50/50 decision over the options Unvoice or Betray
        chlammy=random.randint(1,2)
        decisions= user_decision+"-"+str(chlammy)
        outcomes={"1-1":"\nSeems like I underestimated your relationship, because Chlammy didn't incrimate you. \nHonor between thiefs? Ha! \nLet's see if you survive 2 years in Plum's Dungeon. ","1-2":"\nChlammy didn't say a word, though. You are free, but\nthe weight of your dirty conscience will follow you for the next 10 years as she perishes in Plum's Dungeon! ","2-1":"\nHowever, there is no honor between thiefs. \nChlammy betrayed you and she will taste the freedom \nmeanwhile I see you perishing 10 years in Plum's Dungeon","2-2":"\nBut, there is no honor between thiefs! \nBoth of you are sentenced to spend 7 years in Plum's Dungeon for attempting stealing our treasures."}
        winner={"1-1":"It is a ᴛⷮiͥeͤ! Try playing again to see if you are more lucky.","1-2":"Cͨoͦngrͬaͣᴛⷮuͧlaͣᴛⷮiͥoͦns͛! You are the winner. Try playing again to test your luck.","2-1":"Gaͣmͫeͤ oͦvͮeͤrͬ! Chlammy won and you are in jail. Won't you take revenge?","2-2":"There is no winner. Chlammy and you ended up in Plum's Dungeon. Try again"}
        board_pc_user={"1":"𝙪𝙣𝙫𝙤𝙞𝙘𝙚","2":"𝙗𝙚𝙩𝙧𝙖𝙮"}
        if user_decision=="1":
            print(f"""
            🗯️ 𝓐𝔃𝓻𝓪𝓮𝓵: You decided to {board_pc_user[user_decision]}. {outcomes[decisions]}
            
            {"_"*30}
            {winner[decisions]}
            """)
            break
        elif user_decision=="2":
            print(f"""
            🗯️ 𝓐𝔃𝓻𝓪𝓮𝓵: You decided to {board_pc_user[user_decision]} Chlammy. {outcomes[decisions]}
            
            ║█║▌║█║▌│║▌║▌█║
            {winner[decisions]}
             🅳🅸🆂🅱🅾🅰🆁🅳
            """)
            break
        else:
            print(f"""
            🗯️ 𝓐𝔃𝓻𝓪𝓮𝓵: What is that supposed to mean?
           """)
            continue
else:
    #Creating a loop where PC vs PC play five times
    i=1
    #Scores to declare a winner later
    chlammy_score=0
    fil_score=0
    print(f"""
    🗯️ 𝓐𝔃𝓻𝓪𝓮𝓵: You two, Chlammy and Fil, will be interrogated for five days
    until the truth is revealed...
    """)
    while i<6:
        #Player 1 "thinking" of each resulting escenario as a 25% probability
        chlammy_escenarios={1:1,2:1,3:2,4:2}
        number_chlammy= random.randint(1,4)
        chlammy_by_escenario=int(chlammy_escenarios[number_chlammy])
        #Player 2 making a 50/50 decision over the options Unvoice or Betray
        fil_by_50_50=random.randint(1,2)
        #Combined decision
        fn_decision= str(chlammy_by_escenario)+"-"+str(fil_by_50_50)
        #Possible scenarios outcomes
        escenarios={0:"Honor between thiefs? Ha! \nLet's see if you survive 2 years in Plum's Dungeon.",1:"What an impure follower! Fil betrayed her partner and is free. \nChlammy, 10 years of misery wait for you in Plum's Dungeon.",2:"Your hero betrayed you, Fil.\nYour soul will ache for 10 years in Plum's Dungeon. \nChlammy is now free for your sins!",3:"There is no honor between thiefs! \nYou two will agonize in Plum's Dungeon \nfor 7 years after attempting stealing our treasure."}
        decisions={1:"𝙪𝙣𝙫𝙤𝙞𝙘𝙚",2:"𝙗𝙚𝙩𝙧𝙖𝙮"}
        print(f"""
        ▏𝔻𝕒𝕪 {i}   ▏
        ►Fil decided to {decisions[fil_by_50_50]} and Chlammy to {decisions[chlammy_by_escenario]}.
        """)
        #Summing the wins
        if "1-1" in  fn_decision:
            chlammy_score+=1
            fil_score+=1
        elif "1-2" in fn_decision:
                chlammy_score+=1
        elif "2-1" in fn_decision:
                 fil_score+=1
        else:
            chlammy_score+=0
            fil_score+=0
        i+=1
        continue;
    print(f"""
    {" "*11} 𝔽𝕚𝕟𝕒𝕝 𝕨𝕚𝕟𝕤 {" "*11}
    {"_"*35} 

    𝑪𝒉𝒍𝒂𝒎𝒎𝒚 score: {chlammy_score}
    𝑭𝒊𝒍 score: {fil_score}
    {"_"*35}
    """)
#Defining a winner or trials.
    if chlammy_score == 0 and fil_score == 0:
        print(f"""
        🗯️ 𝓐𝔃𝓻𝓪𝓮𝓵: Everything is said. 
        {escenarios[3]}
         {" "*20}🆃🆁🅸🅰🅻""")
    elif chlammy_score>fil_score:
        print(f"""
        🗯️ 𝓐𝔃𝓻𝓪𝓮𝓵: Everything is said. 
        {escenarios[2]}
        {" "*20}🆆🅸🅽🅽🅴🆁: Chlammy
        """)
    elif chlammy_score<fil_score:
        print(f"""
        🗯️ 𝓐𝔃𝓻𝓪𝓮𝓵: Everything is said. 
        {escenarios[1]}
        {" "*20}🆆🅸🅽🅽🅴🆁: Fil
        """)
    else:
        print(f"""
        🗯️ 𝓐𝔃𝓻𝓪𝓮𝓵: Everything is said. 
        {escenarios[0]}
        {" "*20}🆃🆁🅸🅰🅻
        """)

#
    print( f"""
                ║█║▌║█║▌│║▌║▌█║
                ►The end?
                 🅳🅸🆂🅱🅾🅰🆁🅳""")
```
